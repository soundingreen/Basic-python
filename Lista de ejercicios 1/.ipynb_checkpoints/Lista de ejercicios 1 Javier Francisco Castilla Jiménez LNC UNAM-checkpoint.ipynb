### Ejercicio 2
>2. (9 puntos) A continuación hay una muestra de los precios (en USD) de gasolina estándar por galón en la bahía de San Francisco durante el mes de junio de 1997: 3.88,3.90, 3.93, 3.90, 3.93, 3.96, 3.88, 3.94, 3.96, 3.88, 3.94, 3.99, 3.98 Representa estos datos en: 
(a) Una tabla de frecuencias.
(b) Una gráfica de línea con las frecuencias relativas.
(c) Un histograma.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
precios=[3.88,3.90, 3.93, 3.90, 3.93, 3.96, 3.88, 3.94, 3.96, 3.88, 3.94, 3.99, 3.98] #defina la lista precios que contiene una muestra de los precios (en USD) de gasolina estándar por galón en la bahía de San Francisco durante el mes de junio de 1997
sf={"USD":precios} #creo un diccionario con precios y la clave USD
psgas=pd.DataFrame(sf) # creo un DataFrame que a partir de el diccionario sf
usd=psgas["USD"].unique()#saco los datos unicos de psgas estos datos unicos son mis clases para la tabla de frecuencias #usd es un np.array 
usd.sort()# unique nos devuelve los elementos en orden de aparición por eso ordenamos el array usd usando sort
tabladf= pd.DataFrame(usd, columns=["Clases"])#Creamos el dataframe que contendrá las clases y que posteriormente será mi tabla de frecuencias 
frecuencias = pd.crosstab(index=psgas["USD"],columns="frec_abs") #obtengo las frecuencias absolutas de cada clase utilizando pd.tab #En index indique la variable que quiero contar usando la key USD y en columns especifique el nombre de la columna de salida.
#crosstab genera un DataFrame independiente que almaceno en frecuencias 
afa=frecuencias.values #Creamos una np.array con los valores de las frecuencias 
tabladf["Frec_abs"] = afa #Agregamos la columna de frecuencias absolutas a nuestro dataframe  tabla de frecuencias #crosstab nos arroja la frecuencias ordenadas por eso era importante ordenar primero usd
afr= afa / (len(precios))# calculamos el array de frecuencia relativa #las frecuencias relativas son la misma información que las frecuencias abosolutas pero representada en porcentajes.
tabladf["Frec_rel"]=afr #ahora agregaremos la columna de frecuencia relativa,a nuestra tabla de frecuencias

#ahora agregaremos la columna de frecuencias absolutas acumuladas, en ésta en cada renglón se va sumando el dato de la columna de frecuencias absolutas

#Obtenemos las frecuencias absolutas acumuladas
#Recorremos el array afa para ir creando una nueva lista con las sumas
#Obtener la frecuencia absoluta acumulada
a = []
b = 0
for c in afa:
    b = c + b
    a.append(b)

#Agregamos la nueva columna Frec_abs_acu al Dataframe dat
tabladf["Frec_abs_acu"] = np.array(a)


#ahora agregaremos la columna de Frecuencias relativas acumuladas, en ésta en cada renglón se va sumando el dato de la columna de frecuencias relativas

#Obtenemos las frecuencias relativas acumuladas
#Recorremos el array afr para ir creando una nueva lista con las sumas
#Obtener la Frecuencias relativas acumuladas
x = []
y = 0
for c in afr:
    y = c + y
    x.append(y)

#Agregamos la nueva columna Frec_rel_acum al Dataframe dat
tabladf["Frec_rel_acum"] = np.array(x)

print(tabladf) # imprimimos nuestra tabla de frecuencias nuestra tabla de frecuencias # (a) Una tabla de frecuencias.




```

```python
# forma alternativa para generar la tabla de frecuencias utilizando solo numpy
import numpy as np
datos=[3.88,3.90, 3.93, 3.90, 3.93, 3.96, 3.88, 3.94, 3.96, 3.88, 3.94, 3.99, 3.98]
valores, frecuencia=np.unique(datos, return_counts=True) # contamos los valores unicos y los asignasmos a lista de valores # contamos la frecuencia de cada valor unico y lo asignamos a 
acumulada=np.cumsum(frecuencia)# creamos un array con la frecuencia acumulada
relativa=frecuencia/acumulada[-1] #divimos la frecuencia entre el número elementos que corresponde a acumulada[-1] también corresponde a len(valores)
rel_acumulada=np.cumsum(relativa)#creamos array con la frecuencia relativa acumulada
print("Valores\t Frec\t Rel\t Acu\tRel Acu")# imprimimos encabezados de nuestra tabla de frecuencias
for i in range (len(valores)):#ciclamos para imprimir columnas de la tabla 
    print("{}\t {}\t {:.2f}\t {}\t{:.2f}\t".format(valores[i],frecuencia[i],relativa[i],acumulada[i],rel_acumulada[i]))#damos formato de salida 

    
```

```python
#libro 1 seleccionado Cosmos de Carl Sagan, Fragmento de la introducción.

with open("fragmento 1.txt",mode='r') as libro1: # observemos que el txt tiene que estar en la misma carperta que el script si voy usar esta sintaxis
    contenido1= libro1.readlines() #contenido1 es una list cuyo unico elemento es un string con todo el texto del libro
cosmos=contenido1[0]#aquí ya cerré el archivo txt y tengo su contenido almacenado de un  string que llame cosmos 
print(cosmos)#imprimimos el string
```

```python
#libro 2 seleccionado The Fabric of the cosmos de Brian Greene, fragmento del capítulo 1

with open("fragmento 2.txt",mode='r') as libro2: # observemos que el txt tiene que estar en la misma carperta que el script si voy usar esta sintaxis
    contenido2= libro2.readlines()
fabric=contenido2[0]#aquí ya cerré el archivo txt y tengo su contenido almacenado de un  string que llame fabric
print(fabric)#imprimimos el string
```

```python
for c in ".,()[]-:;¿?": # con este ciclo reemplazo todos los signos de puntuación que aparecen en los textos
    cosmos= cosmos.replace(c,"")
    fabric= fabric.replace(c,"")
```

```python
print(cosmos)# observamos que ambos strings ya no tienen signos de puntuación
print(fabric)
```

```python
import numpy as np
cosmosa=cosmos.split() #con el metodo split creo listas que contienen las palabras en forma de str separadas como elementos de una lista
fabrica=fabric.split()
print(type(cosmosa))# vemos que en efecto son listas y no strings
print(type(fabrica))
print(cosmosa)# vemos que en efecto contienen las palabras separadas como elementos str en forma de lista
print(fabrica)
```

```python
#generamos la tabla de frecuencias utilizando solo numpy
valores, frecuencia=np.unique(cosmosa, return_counts=True) # contamos los valores unicos y los asignasmos a lista de valores # contamos la frecuencia de cada valor unico y lo asignamos a 
orden=np.argsort(frecuencia)
frecuencia=frecuencia[orden]
valores=valores[orden]
acumulada=np.cumsum(frecuencia)# creamos un array con la frecuencia acumulada
relativa=frecuencia/acumulada[-1] #divimos la frecuencia entre el número elementos que corresponde a acumulada[-1] también corresponde a len(valores)
rel_acumulada=np.cumsum(relativa)#creamos array con la frecuencia relativa acumulada
print("Frec\t Rel\t Acu\tRel Acu\t Valores ")# imprimimos encabezados de nuestra tabla de frecuencias
for i in range (len(valores)):#ciclamos para imprimir columnas de la tabla 
    print("{}\t {:.2f}\t {}\t{:.2f}\t{}\t".format(frecuencia[i],relativa[i],acumulada[i],rel_acumulada[i],valores[i]))#damos formato de salida 
```

```python
# grafica para apoyar la opinión de que las distribuciones de frecuencias son similares
plt.plot(valores,relativa)
plt.show()
```

4.(3 puntos) A partir del archivo Data200g0.csv, crea un histograma utilizando la
frecuencia relativa.
 >**(a)¿Cuánto vale el área definida por el histograma? >Cuánto valdría si usásemos la
frecuencia absoluta (en vez de la relativa)?**
#### R: vale 1 y si usaramos la frecuencia absoluta  en vez de la relativa valdría 200

 >**(b)¿Qué ancho de ventana utilizas para el histograma y por qué?**
#### R: 0.038632432827416394 que es el que corresponde a utilizar 30 bins. Elegí 30 bins porque me permitía observar con más claridad la frecuencia, es decir con menos bins parecía que algunas clases tenían mayor frecuencia pero no podía observar como hay intervalos de clase vacíos

```python
import numpy as np
import matplotlib.pyplot as plt
data200g0= np.loadtxt("Data200g0.csv")#cargamos los datos en un np array usando np.loadtxt()
counts, bins=np.histogram(data200g0,bins=30) #count guarda un array con la frecuencia y bins un array con los bordes de bins
counts=counts/counts.sum() #convertimos la frecuencia frecuencia relativa
print(counts.sum())# imprimos la suma de las frecuecias relativas solo para corroborar que suman 1

plt.hist(bins[:-1],bins,weights=counts) # siguiendo la documentación creamos un histograma usando plt.hist a partir de datos en obtenidos vía np.histogram

w=bins[2]-bins[1]# el ancho de bin
print(w)#imprimos el ancho de bin
```

### Ejercicio 7
7. (35 puntos) Grafica la curva de probabilidad acumulativa para los datos de cada archivo. Para calcular estas curvas es necesario tener los histogramas correspondientes.
>**(a)Compara las curvas obtenidas de los datos de Data50g0.csv y Data500g0.csv con la obtenida con Data200g0.csv.>Qué diferencias observas?**
#### R: es muy notorio que mientras más datos hay en la muestra, la curva está más suavizada  y tiene menos irregularidades,menos picos. Es decir mientras más datos más se parece la curva a una curva de distribución ideal y continua
>**(b)Compara las curvas obtenidas de los datos Data200g2.csv, Data200P2.csv y Data200P4.csv con la obtenida con Data200g0.csv.** **¿Qué diferencias observas?**
#### R: las gráficas de Data200g2.csv, Data200P2.csv y Data200P4.csv tienen muchas irregularidades no son realmente curvas, tienen picos y en particular  Data200P2.csv y Data200P4.csv son literalmente escalonadas, rectilineas más que curvas, todas estas son diferencias con respecto a Data200g0.csv la cuál es una curva bastante más suave y por así decirlo ideal.
>**¿A qué piensas que se deben esas diferencias?**
#### R: parece ser el efecto de tener clases vacias de forma que la distribución es interrumpida por así decirlo, o sea mientras más clases sin datos hay más irregularidades tiene la curva y menos curva es. Parece que mientras mientras más distancia hay entre una clase con datos y la siguiente más acentuado es este patrón rectilineo de escalera en la "curva". En particular también con respecto a Data200g0.csv se debe a que son datos provenientes de experimentos distintos y no al número de datos.

```python
###Data200g0.csv###
import numpy as np
import matplotlib.pyplot as plt
data200g0= np.loadtxt("Data200g0.csv")#cargamos los datos en un np array usando np.loadtxt()
counts, bins=np.histogram(data200g0,bins=30) #count guarda un array con la frecuencia y bins un array con los bordes de bins
counts=counts/counts.sum() #convertimos la frecuencia frecuencia relativa dividiendo entre el numero de elementos en la muestra
prob_acum=counts.cumsum()#creamos un array con los valores de la frecuencia relativa acumulada es decir de la probabilidad acumulada usando .cumsum
#antes de hacer un plot de lineas tengo un problema 
print(len(counts)) # counts y bins tienen dimensiones distintas 30 y 31 elementos respectivamente lo cuál es un problema
print(len(bins))# y es que bins son las cotas inferior y superior de cada bin (intervalo de clase)
#entonces tengo que elgir un solo punto para representar la clase y este punto puede ser la cota inferior 
#la cota superior o un punto medio entre las cotas que represente a la clase 
#voy a optar por un punto medio 
#para obtener este punto medio voy a calcular el valor promedio entre las cotas inferior y superior
clases=((bins[0:30]+bins[1:31])/2) # este array tiene el valor promedio entre la cota inferior y la cota superior
 #observemos que para poder reciclar el código con otros datos se puede usar esta sintaxis más general 
#entonces ya puedo crear un plot de la probabilidad acumulada usando los array clases y prob_acum!!!!
plt.plot(clases,prob_acum)# esta es mi curva de distribución acumulada
plt.show()


```

```python
###Data50g0.csv###
data50g0= np.loadtxt("Data50g0.csv")#cargamos los datos en un np array usando np.loadtxt()
counts, bins=np.histogram(data50g0,bins=20) #count guarda un array con la frecuencia y bins un array con los bordes de bins
counts=counts/counts.sum() #convertimos la frecuencia frecuencia relativa dividiendo entre el numero de elementos en la muestra
prob_acum=counts.cumsum()#creamos un array con los valores de la frecuencia relativa acumulada es decir de la probabilidad acumulada usando .cumsum
clases=((bins[0:(len(bins)-1)]+bins[1:len(bins)])/2) # calcular el valor promedio entre las cotas inferior y superior
plt.plot(clases,prob_acum)# esta es mi curva de distribución acumulada
```
