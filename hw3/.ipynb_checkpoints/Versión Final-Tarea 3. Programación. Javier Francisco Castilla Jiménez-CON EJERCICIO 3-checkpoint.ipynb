```python
# x=5 
lista_error=[] # lista para almacenar los valores de error a graficar
ordendaprox=[]# lista para almacenar los valores de orden a graficar
real=math.exp(5)# el valor real que nos sirve de referencia para calcular el error
for n in range (11):# n desde n=0 hasta n=10
    aproximado=Javier.aproxex(5,n) #calculamos el valor aproximado para orden n
    error=np.abs((real-aproximado)/real)*100 # error relativo porcentual
    ordendaprox.append(n) #.agregamos el valor aproximado calculado nuestra lista de valores aproximados
    lista_error.append(error) # agregamos el valor de error a la lista de error
    
    # print('n', n, 'valor aproximado',aproximado,'error',error)#  esto no se pide pero yo lo quise hacer para ver las tabulaciones antes de gráficar
    
```

```python
import numpy as np
Datos_r=np.loadtxt("Datos_r.csv",delimiter=',')

#generemos 2 puntos aleatorios:
minimos=np.max(Datos_r,axis=0) #usamos np.max y especificamos axis=0 para indicar que queremos los maximos de las columnas. observemos que evitamos el uso de ciclo for para optimizar el código.
maximos=np.min(Datos_r,axis=0) #usamos np.min y especificamos axis=0 para indicar que queremos los minimos de las columnas.
punto1=np.random.uniform(minimos,maximos) #creamos dos puntos aleatorios entre nuestro valores extremos
punto2=np.random.uniform(minimos,maximos) 

print('punto1',punto1) # imprimos las coordenadas de nuestros puntos solo para saber con que punto aleatorios estamos trabajando 
print('punto2',punto2)
```

Recordemos que  Dado un vector de un espacio vectorial euclídeo, la norma de un vector se define como la distancia euclídea (en línea recta) entre dos puntos A y B que delimitan dicho vector. De hecho, en un espacio euclídeo la norma de un vector coincide precisamente con el módulo del vector $${\displaystyle {\vec {AB}}}$$

En el caso general de un espacio euclídeo de n dimensiones se tiene:
$${\displaystyle \|{\vec {AB}}\|={\sqrt {(b_{1}-a_{1})^{2}+(b_{2}-a_{2})^{2}+...+(b_{n}-a_{n})^{2}}}}$$ 

siendo $${\displaystyle {\vec {OA}}=(a_{1},a_{2},...,a_{n})}$$
y
$${\displaystyle {\vec {OB}}=(b_{1},b_{2},...,b_{n})}$$



fuente:https://es.wikipedia.org/wiki/Norma_vectorial

```python
## calculando las distancias a nuestro puntos aleatorios

distancia1=np.sqrt((np.sum((Datos_r-punto1)**2,axis=1))) #solo estamos aplicando la formula mencionada arriba pero para un caso n=8 dimensiones
distancia2=np.sqrt((np.sum((Datos_r-punto2)**2,axis=1)))
distancias=np.column_stack((distancia1,distancia2)) #distancia 1 y distancia 2 son columnas de un nuevo array llamado distancias
print(distancias.shape) # matriz de 2 columnas y n=37799 renglones
print(distancias)
```

```python
igual=distancias[:,0]==distancias[:,1] #este array booleano tiene True si las distancias a ambos puntos son iguales
cercad1=distancias[:,0]<distancias[:,1] #este array booleano tien True si el elemento está más cerca del punto 1
cerca=[] #declaro una lista vacía donde voy almacenar las clasificaciones
for i in range(distancias.shape[0]): #iteramos en el rango de los renglones de nuestro arreglo de distancias
    if igual[i]==True: # si la distancia a ambos puntos es igual entonces:
        moneda=np.random.rand(1) #lanzo una moneda (un número aleatorio entre 0 y 1)
        if moneda>=0.5: #si mi número aleatorio es mayor o igual que 0.5
            cerca.append(1)# entonces voy decir que está más cerca de del punto 1
        else: # si mi número aleatorio es menor que 0.5 
            cerca.append(2) #entonces diré que está más cerca del punto 2 (agrego un 2 a mi lista de clasificaciones)
    elif cercad1[i]==True: # si la distancia a ambos puntos no es igual reviso si está más cerca del punto 1
        cerca.append(1) # si está más cerca del punto 1 entonces agrego un uno a mi lista de clasificaciones
    else: # sino se cumplio nada de lo anterior es porque el punto está más cerca del punto 2
        cerca.append(2) #si está más cerca del punto 2 entonces agrego un 2 a mi lista de clasificaciones
cercania=np.array(cerca)# convertimos la lista con las cercanias en un array


np.savetxt('clasificaciones1.csv',cercania,delimiter=',') #guardamos las clasificaciones como csv
np.save('clasificaciones1.npy',cercania)#guardamos las clasificaciones como npy

# guardamos en pickle:
import pickle    #Carga la librería necesitada 
pickle_out= open('clasificaciones.pickle','wb')
pickle.dump(cercania, pickle_out)
pickle_out.close()

#verificación de que guarde correctamente en pickle:
#pickle_in=open('clasificaciones.pickle','rb') 
#example=pickle.load(pickle_in)
#print(example)
```

```python
import matplotlib.pyplot as plt
import numpy as np
### aún sin terminar lo entregaré mañana
p2c=np.loadtxt("Datos_r.csv",delimiter=',',usecols=(0,1))# cargamos solo las primeras dos columnas
maxi=np.max(p2c,axis=0) #usamos np.max y especificamos axis=0 para indicar que queremos los maximos de las columnas. observemos que evitamos el uso de ciclo for para optimizar el código.
mini=np.min(p2c,axis=0) #usamos np.min y especificamos axis=0 para indicar que queremos los minimos de las columnas.
V1=np.linspace(mini[0],maxi[0],200) # generamos V1 entre el máximo y el mínimo de la colulmna 1 (indice 0)
V2=np.linspace(mini[1],maxi[1],200) # generamos V2 entre el máximo y el mínimo de la colulmna 2 (indice 1)

conteos=np.zeros((200,200),dtype=np.int32) # matriz conteos de 200x200 , es una matriz "vacía"

for renglon in range(199):
    for columna in range(199):
        condicion1=np.logical_and((p2c[:,0]>=V1[renglon]),(p2c[:,0]<V1[renglon+1]))# matriz boolena de los elementos que cumplen que son mayores o iguales que el elemento 1 de V1 pero menores que el primer elemento de V1.Nos movemos sobre los elementos de V1 para los renglones y V2 para las columnas.
        condicion2=np.logical_and((p2c[:,1]>=V2[columna]),(p2c[:,1]<V2[columna+1]))# matriz boolena de los elementos que cumplen que son mayores o iguales que el elemento 1 de V2 y menores que el elemento 1 de V2. Nos movemos sobre los elementos de V2 para las columnas.
        
        conteos[199-renglon,columna]=np.sum(np.logical_and(condicion1,condicion2)) # el conteo de todos los elementos que cumplen que son mayores o iguales que el elemento 1 de V1 pero menores que el primer elemento de V1 y, ademas que son mayores o iguales que el elemento 1 de V2 y menores que el elemento 2 de V2
        
plt.imshow(conteos) #utilizando imshow mostramos la matriz generada
```

```python
import numpy as np
# Cargué el archivo RR032157_007_Psyc.csv. Este tiene como separador \t (tabulador) y tiene cabeceras
cabeceras=np.loadtxt("RR032157_007_Psyc.csv", delimiter="\t", max_rows=1, dtype=str)
datos=np.loadtxt("RR032157_007_Psyc.csv", delimiter="\t", dtype=np.float32, skiprows=1)
print('los encabezados son:',cabeceras) #imprimimos las cabeceras para saber en que columna están las respuestas
print("\n La dimensión de los datos es: ", datos.shape) #imprimimos la dimension de los datos para tener idea de con que estamos lidiando
# a partir de shape ya sabemos que nuestra cantidad total de ensayos es:
ensayos=datos.shape[0] # almacenamos la cantidad total de ensayos que corresponde a la cantidad total de respuestas 
print('la cantidad total de ensayos es:', ensayos)
```

```python
#En base a las cabeceras sabemos que las respuestas están en la colulmna tres 
# Utilizando el comando np.unique, determinamos las posibles respuestas en la columa 3 del archivo RR032157_007_Psyc.csv
tres=np.loadtxt("RR032157_007_Psyc.csv", delimiter="\t", skiprows=1,usecols=(2))
print(np.unique(tres))
print('las respuestas únicas son',np.unique(datos[:,2]))
respuestas_posibles=np.unique(tres) # respuestas_posible almacena las respuestas posibles respuestas, solo son 3 
#veamos las amplitudes (columnas 6 y 7):
amplitudes=np.stack((datos[:,5],datos[:,6]),axis=-1)
print(amplitudes)
print(amplitudes.shape)
print('---------------------------------------------------------------------------')
ampunic=np.unique(amplitudes,axis=0)# pares de amplitudes únicas 
print('los pares de amplitudes son:', ampunic)
print(ampunic.shape)
print('---------------------------------------------------------------------------')
respuestas=datos[:,2]# solo quería visualizar mis datos
print(respuestas)
print(respuestas.shape)
print('el primer elemento de la colulmna de respuestas es:',respuestas[0])
```

```python
indices=[] #declaro listas vacias
ndanswers=[]
for i in range (ampunic.shape[0]): # en el rango de la cantidad pares de amplitudes unicos
    a=np.apply_along_axis(sum,1,(amplitudes==ampunic[i])) #los renglones que al sumar tengan un dos son donde está ese par de amplitudes 
    coordenadas=list(np.where(a==2)[0]) #las coordenadas que cumple que tienen un dos son las de aquellos renglones donde aparece ese par de amplitudes
    respuestasxpar=len(coordenadas) # este es el numero de respuestas dado un par de amplitudes único
    ndanswers.append(respuestasxpar) # cada elemento de esta lista es el número de respuestas correspondientes cada un par único de amplitudes 
    indices.append(coordenadas) # cada sublista de indices tiene las coordenadas de la fila en que se respondión un par especifico de coordenadas unicas

print(len(ndanswers)) #el tamáño de ndanswers el mismo que el de ampunic porque se corresponden uno a uno
print(len(indices)) # el tamaño de indices es el mismo que el de ndanswers y ampunic porque se corresponden uno a uno
print(indices)
```

```python
# en el orden de los pares respuestas en el orden de 
rorden=[]
for i in range (ampunic.shape[0]):#itera la cantidad pares de amplitudes unicos es decir son 11 ciclos
    for index in indices[i]: #itera sobre los elementos de cada sublista en indices 
        rorden.append(respuestas[index])#los elementos de la lista rorden corresponden  uno a uno a las respuestas de los indices pero sin las sublistas
print(rorden)
print(len(rorden))
```

```python
# tamaño de particiones
tam=[]
for i in range (ampunic.shape[0]):#itera la cantidad pares de amplitudes unicos es decir son 11 ciclos
    tam.append(len(indices[i]))# tam es el tamaño de elementos de 
print(tam) 
```

```python
from itertools import islice # usamos itertools para dividir la lista rorden en sublistas de tamaños dados (tam)
entrada = rorden # lista a particionar
tamaños = tam # lista de tamaños de las particiones
# Usando islice:
inputt = iter(entrada) 
salida = [list(islice(inputt, elem))for elem in tamaños]
print(salida) # en esta lista de salidas cada sublista son los respuestas a que de cada un par 
#en esta lista en cada sublista los indices tiene las coordenadas de la fila en que se corresponde con
# la de un par especifico de coordenadas unicas en el array ampunic (de amplitudes únicas)
```
