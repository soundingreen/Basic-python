```python
import matplotlib.pyplot as plt
import numpy as np
### a√∫n sin terminar lo entregar√© ma√±ana
p2c=np.loadtxt("Datos_r.csv",delimiter=',',usecols=(0,1))# cargamos solo las primeras dos columnas
maxi=np.max(p2c,axis=0) #usamos np.max y especificamos axis=0 para indicar que queremos los maximos de las columnas. observemos que evitamos el uso de ciclo for para optimizar el c√≥digo.
mini=np.min(p2c,axis=0) #usamos np.min y especificamos axis=0 para indicar que queremos los minimos de las columnas.
V1=np.linspace(mini[0],maxi[0],200) # generamos V1 entre el m√°ximo y el m√≠nimo de la colulmna 1 (indice 0)
V2=np.linspace(mini[1],maxi[1],200) # generamos V2 entre el m√°ximo y el m√≠nimo de la colulmna 2 (indice 1)

conteos=np.zeros((200,200),dtype=np.int32) # matriz conteos de 200x200 , es una matriz "vac√≠a"

for renglon in range(199):
    for columna in range(199):
        condicion1=np.logical_and((p2c[:,0]>=V1[renglon]),(p2c[:,0]<V1[renglon+1]))# matriz boolena de los elementos que cumplen que son mayores o iguales que el elemento 1 de V1 pero menores que el primer elemento de V1.Nos movemos sobre los elementos de V1 para los renglones y V2 para las columnas.
        condicion2=np.logical_and((p2c[:,1]>=V2[columna]),(p2c[:,1]<V2[columna+1]))# matriz boolena de los elementos que cumplen que son mayores o iguales que el elemento 1 de V2 y menores que el elemento 1 de V2. Nos movemos sobre los elementos de V2 para las columnas.
        
        conteos[199-renglon,columna]=np.sum(np.logical_and(condicion1,condicion2)) # el conteo de todos los elementos que cumplen que son mayores o iguales que el elemento 1 de V1 pero menores que el primer elemento de V1 y, ademas que son mayores o iguales que el elemento 1 de V2 y menores que el elemento 2 de V2
        
plt.imshow(conteos) #utilizando imshow mostramos la matriz generada
```
