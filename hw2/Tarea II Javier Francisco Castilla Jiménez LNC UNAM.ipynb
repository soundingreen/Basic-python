```python
#4. (5 puntos) Realice una función que calcule el area de la función seno entre 0 y pi utilizando el método de MonteCarlo. 
#La función debe recibir el número de iteraciones como argumento de entrada y debe devolver la aproximación. 

def areasin(iteraciones,paso=100): 
    """
    Está función fue creadada para aproximar el area de la función seno entre 0 y pi utilizando el método de MonteCarlo.
    La función recibe el número de iteraciones que queremos realizar y los pasos a los que cambiamos la cantidad 
    de numeros aleatorios que se ocupan en cada iteración, el parámetro paso también corresponde a la cantidad de números aleatorios
    que se genera en la primera iteración que van aumentando según el mismo valor de paso. La función devuelve un array con la aproximación
    que corresponde a cada iteración.Por ejemplo si cálculamos areasin(250) nos regresará un numpy array de 250 elementos correspondientes a 
    a cada iteración, es decir el elemento uno corresponde a la iteración 1, el elemento dos corresponde a la iteración 2 y así 
    respectivamente. Si calculamos areasin(250,20) entonces vamos a iterar 250 veces pero ahora en vez de empezar la primera iteración
    con 100 números aleatorios vamos a empezar con veinte números y cada iteración subsecuente iremos agregando 20 números 
    aleatorios más con respecto a la iteración anterior.
    <parametros>
    iteraciones: integer>=1
    paso: integer>=1 opcional el valor default es 100
    <devuelve>
    np.array que contiene los valores aprovimados correspondiente a cada iteración
    
    
    """
    
    final=(iteraciones*paso)+paso #la cota superior de mi rango es mi numero de iteraciones multiplicada por el tamaño de los pasos + un paso
    inicio=paso # por conveniecia inicio en el tamaño de mis paso en el rango
    import math 
    import numpy as np
    lista=[]
    for i in range (inicio,final,paso):
        ejex=(np.random.rand(i))*math.pi #numpy array de i numeros aleatorios en [0,mt.pi)
        ejey=np.random.rand(i) #numpy array de i numeros aleatorios en [0,1)
        ydelsin=np.sin(ejex) #matriz de valores sin(x) de las x del np array ejex #los valores de y=sin(x) delimitan si un par (x,y) coordenado está fuera 
        #o dentro del area que buscamos 
        dentro=np.sum(ejey<=ydelsin) # si ejey<=ydelsin se cumple cumple que el par coordenado está dentro del area si no 
        rectangulo=math.pi*1 # base por altura
        area=rectangulo*(dentro/i) #i es la cantidad de puntos totales según la iteración
        lista.append(area) # agregamos la aproximación a una lista
    areas=np.array(lista)#convertimos la lista final en un array
    return(areas)# la funcion devuelve un array que tiene las
    

areasin(250)# probamos la función 
```

```python
def conversion(G):
    """
    Esta función convierte de grados Farenheit a grados Celsius y viceversa, es decir, también convierte de grados Celsius 
    a grados Farenheit. la función tiene un solo parámetro el cuál es obligatorio, G, G es el numero que representa los grados
    que se quieren convertir.Una vez ejecutada la función le pregunta al usuario si quiere convertir a grados Farenheit (ºC->ºF)
    o a grados Celsius (ºF->ºC). Si quiere convertir a grados Farenheit (ºC->ºF) debe introducir F, si quiere convertir a grados
    Celsius (ºF->ºC) debe introducir C. Si el usuario no elige una opción adecuada (C o F) la función le seguirá preguntando hasta
    que se seleccione una opción válida.
    la función esta programada para marcar error si se introducen temperaturas menores al cero absoluto por lo tanto si se va a 
    convertir a Celsius no se debe introducir temperaturas menores a los -459.67 ºF , y si se va convertir a Farenheit no se 
    deben introducir temperaturas menores a los -273.15 ºC.

    <parámetros>
    G es un parámetro oblligatorio debe ser un float o un int
    <devuelve>
    un float que es la temperatura convertida
    Descripción de la función. Como debe ser usada,
    que parámetros acepta y que devuelve
    """
    assert(type(G)==float or type(G)==int), 'el parametro introducido debe de clase float o int'
    indicador=False #indicador es False al iniciar
    x=input('¿quieres convertir a grados farenheit? (introduce F) o ¿quieres convertir a grados celsius? (introduce C) :')
    if x=='F': # Si voy a convertir a Farenheit es porque introduje grados Celsius en G para Celsius el cero absoluto es −273.15 °C
        assert(G>=273.15*-1),'el número introducido debe ser mayor que el cero absoluto' 
        indicador=True #si el usuario introduce una respuesta válida indicador cambia a True
        return (G*9/5) +32 #conversion de celsius a farenheit
    elif x=='C': # Si voy a convertir a Celsius es porque introduje grados Farenheit en G para Farenheit el cero absoluto es −459.67 °F
        assert(G>=459.67*-1),'el número introducido debe ser mayor que el cero absoluto' 
        indicador=True
        return ((G-32)/(9/5))#conversion de farenheit a celsius
    else:
        indicador=False
    while indicador==False: # si indicador es False seguir preguntando hasta que el usuario introduzca una respuesta válida lo cuál cambia indicador a True 
        x=input('¿quieres convertir a grados farenheit? o ¿quieres convertir a grados celsius? ') 
        if x=='F':
            assert(G>=273.15*-1),'el número introducido debe ser mayor que el cero absoluto' 
            indicador=True
        elif x=='C':
            assert(G>=459.67*-1),'el número introducido debe ser mayor que el cero absoluto'
            indicador=True
    if x=='F':# si salimos del ciclo while es porque el porque el usuario introdujo una respuesta válida y podemos realizar los cálculos correspondientes
        return (G*9/5) +32
    elif x=='C':
        return ((G-32)/(9/5))
    
```
