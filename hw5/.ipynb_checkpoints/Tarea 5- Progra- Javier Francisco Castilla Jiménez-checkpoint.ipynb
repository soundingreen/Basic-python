```python
class CalDifNum():
    def __init__(self, señal,tiempo,a,b):
        self.señal=señal# atributo señal
        self.tiempo=tiempo #atributo tiempo
        self.a=a# limite inferior de integración
        self.b=b# limite superior de integración
    def itrapecio(self): # método para calcular la integral con el métódo del trapecio
        times=self.tiempo
        signal=self.señal 
        a=self.a # este son los valores de a para sacar la integral debajo de toda la señal
        b=self.b # este son los valores de b para sacar la integral debajo de toda la señal
        dom=len(times)# el número de elmentos que tienen mis tiempos y la señal
        i_inicial=int(np.where(times==a)[0])#indice que corresponde al limite a
        i_final=int(np.where(times==b)[0]) #indice que corresponde al limite b
        n=dom-(np.abs(i_inicial)+np.abs(dom-i_final)) # n es igual a mi número de tiempos
        suma=np.sum(signal[1:n])#sumatoria de valores entre los indices 1 y n
        I=(((b-a))*(signal[i_inicial]+2*suma+signal[i_final]))/(2*n)#calculamos la integral
        return I
```

```python
class CalDifNum():
    def __init__(self, señal,tiempo,a,b):
        self.señal=señal# atributo señal
        self.tiempo=tiempo #atributo tiempo
        self.a=a# limite inferior de integración
        self.b=b# limite superior de integración
    def itrapecio(self): # método para calcular la integral con el métódo del trapecio
        times=self.tiempo
        signal=self.señal 
        a=self.a # este son los valores de a para sacar la integral debajo de toda la señal
        b=self.b # este son los valores de b para sacar la integral debajo de toda la señal
        dom=len(times)# el número de elmentos que tienen mis tiempos y la señal
        i_inicial=int(np.where(times==a)[0])#indice que corresponde al limite a
        i_final=int(np.where(times==b)[0]) #indice que corresponde al limite b
        n=dom-(np.abs(i_inicial)+np.abs(dom-i_final)) # n es igual a mi número de tiempos
        suma=np.sum(signal[1:n])#sumatoria de valores entre los indices 1 y n
        I=(((b-a))*(signal[i_inicial]+2*suma+signal[i_final]))/(2*n)#calculamos la integral
        return I
    def simpson(self): # método para calcular la integral con el métódo del trapecio
        times=self.tiempo
        signal=self.señal 
        a=self.a # este son los valores de a para sacar la integral debajo de toda la señal
        b=self.b # este son los valores de b para sacar la integral debajo de toda la señal
        dom=len(times)# el número de elmentos que tienen mis tiempos y la señal
        i_inicial=int(np.where(times==a)[0])#indice que corresponde al limite a
        i_final=int(np.where(times==b)[0]) #indice que corresponde al limite b
        n=dom-(np.abs(i_inicial)+np.abs(dom-i_final)) # n es igual a mi número de tiempos
        sum1=np.sum(signal[1:-1:2])
        sum2=np.sum(signal[2:-1:2])
        I=(((b-a)/(n))*(signal[i_inicial]+4*sum1+2*sum2+signal[i_final]))/(3)#calculamos la integral
        return I # devuelve el valor aproximado de la integral
    def montecarlo(self):# metodo para cálcular la integral con el método de montecarlo
        times=self.tiempo
        signal=self.señal
        a=self.a # este son los valores de a para sacar la integral debajo de toda la señal
        b=self.b # este son los valores de b para sacar la integral debajo de toda la señal
        dom=len(times)# el número total de elmentos que tienen mis tiempos y la señal
        i_inicial=int(np.where(times==a)[0])#indice que corresponde al limite a
        i_final=int(np.where(times==b)[0]) #indice que corresponde al limite b
        n=dom-(np.abs(i_inicial)+np.abs(dom-i_final)) # n es igual a mi número de tiempos
        maxy=np.max(signal[i_inicial:i_final])
        ejex=(np.random.rand(n))*(b-a)
        ejey=np.random.rand(n)*maxy
        dentro=np.sum(ejey<=signal[i_inicial:i_final]) # si ejey<=signal[i_inicial:i_final] se cumple cumple que el par coordenado está dentro del area si no 
        rectangulo=(b-a)*maxy #area donde genere puntos aleatorios
        area=rectangulo*(dentro/n)
        return area
```

```python
class nrn(list):
    
    """
    La clase nrn está hecha para almcenar registros de espigas correspondientes a un solo ensayo. incluye los métodos
    tasa, para calcular la tasa de disparo, isi para obtener el intervalo interespiga y binary para mostrar la binarización
    del registro.
    El método tasa requier los parametros obligatorios, start (donde empieza a contar), stop(donde termina de contar), paso
    y ventana, estos ultimos dos parametros deben expresarse en segundos. el argumento opcional det sirve para obtener también 
    los valores de las ventanas deterministicas para gráficar.Por default det=False 
    El método isi devuelve un array con los intervalos interespiga.
    El método binary realiza la binarización y tiene los argumentos opcionales , start y end, que por default valen uno y
    diez respectivamente.
    Los datos que se introducen a esta clase no puede haber más de 1 espiga por ms de lo contrario marcara error, esto es por
    el periodo refractario.
    
    """
    
    def __init__(self,espigas): # inicializo mi clase # con el atributo espigas
        import numpy as np # importo numpy para poder hacer los calculos con fluidez
        import Javier as mi # importo mi librería personal 
        super().__init__()# super para heredar facilmente 
        self.__ensayo=np.array(espigas)# convierto mi ensayo a un numpy array para operar con el 
        self.__i_final=self.__ensayo[1:]# datos superiores
        self.__i_inicial=self.__ensayo[0:-1] # datos inferiores
        self.__inter=self.__i_final-self.__i_inicial# intervalo interespiga
        if (np.sum((self.__inter)<0.001))>=1: # error si hay más de un disparo un disparo en un intervalo de 1ms 
            raise NameError('no puede haber mas de una espiga en un intervalo de 1 ms')
        else:
            self.espigas=espigas# mi 
            
    def tasa(self,start,stop,paso,ventana,det=False): # metodo para calcular la tasa de disparo 
        bsu,nventanas=mi.bordesup(start,stop,paso,ventana,nv=True)# calculo los bordes superiores de mi
        tasa=mi.tasa(self.__ensayo,nventanas,paso,ventana)# calcula la tasa de disparo
        if det==True: # si el argumento
            return bsu,tasa
        else:
            return tasa
        
    def isi(self): #método para intervalo interespigas
        isi=self.__inter
        return isi #devuelve arreglo de intervalos interespigas
    
    def binary(self,start=0,stop=10):# metodo de binarización
        t=self.tasa(start,stop,0.001,0.001)# calculo a forma de tasa
        binario=(t!=0)*1 # binarización
        return binario # devuelve la binarización
    
    def __add__(self,value):# la suma debe marcar error
        raise NameError('no puede sumar a sus espigas')
        
    def __mul__(self,numero):# la multiplicación debe marcar error
        raise NameError('no puede multiplicar a sus espigas')
        
    def __truediv__(self,value):# la división debe marcar error
        raise NameError('no puede dividir a sus espigas')
        
    def __sub__(self,value):#la resta debe marcar error
        raise NameError('no puede restar a sus espigas')
        
```
