```python
# prueba de código par luego realizar función que actualice los datos artificiales
clasificacion=Javier.ejercicio2(Datos_r,2)
arreglo=Datos_r
#######
tiposclas=np.unique(clasificacion) #los valores unicos en los que clasificamos
centroid=[] # declaro lista vacía par guardar
for clase in tiposclas: #iteramos cada uno de estas clasificaciones únicas
    i_subclas=np.where(clasificacion==clase)[0] #los indices de clasificaciones donde hay ese tipo de clasificación particular que corresponde a la varible clase
    print(i_subclas) # imprimi solo para ver que estaba haciendo lo que yo esperaba
    nuevos=arreglo[i_subclas] #aquí hago un slice de los renglones clasificados en clase 
    c_mean=np.mean(nuevos,axis=0) # saco el promedio por renglones del slice llamado nuevos
    print(c_mean) # imprimi solo para ver que estaba haciendo lo que yo esperaba
    centroid.append(c_mean)#aquí agrego cada vector centroide formado con los promedios por columna a una lista de centroides llamada centroid
centroides=np.array(centroid) #convierto mi lista de centroides en un array donde cada elemento es un centoide
print('_____________________________________________')
print('centroides',centroides)
```

**¿Cuántas iteraciones necesita para obtener un distancia del orden de 1e−6?**
> Varía de acuerdo a las semillas random generadas pero en este caso parcen ser alrededor de 5

```python
clasificacion,seeds=Javier.ejercicio2mod1(Datos_r,2,return_semillas=True)# ejecutamos el ejercicio uno y generamos la primera clasificacion y un arreglo con las semillas
print('seeds\n',seeds)
arreglo=Datos_r # llamo mi arreglo de datos
numditer=[] # guardo el conteo de las iteraciones en esta lista 
previous=seeds# centroide previo para la primera iteración el centroide previo son la semillas
distancias=[] #guardo las distancias entre un centroide y el previo
for iteracion in range (1,31):# 30 iteraciones
    numditer.append(iteracion)# voy contando la iteraciones y agregandolas en un lista
    centroides=Javier.newcentroid(arreglo,clasificacion) # genero los centroides pormediando columnas
    clasificacion=Javier.clasificacion_vertar3(arreglo,centroides)# genero clasificacion en base a mis centroides
    distanciasC=Javier.norma(centroides,previous) # calculo la distancia entre un centroide y el centroide de la it anterior
    distancias.append(distanciasC)# agrego la distancia entre centroides por iteracion a la lista de distancias
    previous=centroides # actualizo el valor del centroide previo
distancias=np.array(distancias) # la lista de distancia entre centroides la convierto en un array
print('distancias entre centroides',distancias) # imprimo solo para ver 
print(len(distancias))
print('lista de iteraciones',numditer)
```

```python
plt.plot(numditer,distancias[:,0])
plt.plot(numditer,distancias[:,1])
fontdict_title = {'family': 'serif','color':  'darkblue','weight': 'normal','size': 16,} # formato para el título
plt.title("Tasa de disparo Neurona DPC1\n Clase 1", fontdict_title) #título de la gráfica
plt.xlabel(" # de iteraciones ") # Configuramos la etiqueta del eje X
plt.ylabel(" distancia entre centroides") # Configuramos la etiqueta del eje Y
plt.show()
```
